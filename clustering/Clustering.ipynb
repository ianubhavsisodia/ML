```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Population_Data.csv")
data.info()
```

```python
data.head()
```

```python
numeric = ["Indians",'Foreigners',
          'Indian_Male','Indian_Female',
          'Foreigners_Female',
          'Foreigners_Male','Total Population']
```

```python
def cleaner(z):
    return z.replace(',','')

cleaner('100,234')
```

```python
for i in data[numeric]:
    data[i] = data[i].apply(cleaner)

data.head()
```

```python
data.info()
```

```python
data[numeric] = data[numeric].apply(pd.to_numeric)
data.info()
```

```python
data[['Indians','Foreigners']].sum().sum() - data['Total Population'].sum()
```

```python
data[['Indian_Male','Foreigners_Male','Indian_Female','Foreigners_Female']].sum().sum() - data['Total Population'].sum()
```

```python
MF_sum = data['Indian_Male'] + data['Indian_Female'] + data['Foreigners_Male'] + data['Foreigners_Female']
data['other'] = data['Total Population'] - MF_sum
data.head()
```

```python
data['Region'].nunique(), data['Office Location Id'].nunique()
```

```python
data1 = data.drop(columns = ['Region', 'Office Location Id', 'Total Population'])
data1.head()
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
columns = data1.columns
data1 = norm.fit_transform(data1)

data1 = pd.DataFrame(data1, columns = columns)
data1.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data1)
pred = kmeans.predict(data1)
```

```python
pred, len(pred)
```

```python
pd.Series(pred).value_counts()
```

```python
kmeans.inertia_
```

```python
kmeans.score(data1)
```

```python
# to check optimal no. of clusters
SSE = []
for cluster in range(1,10):
    kmeans = KMeans(n_clusters = cluster)
    kmeans.fit(data1)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster':range(1,10), 'SSE':SSE})
```

```python
plt.figure(figsize = (12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker = 'o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(data1)
pred = kmeans.predict(data1)
```

```python
data1['cluster'] = pred
```

```python
# str_ and str_y represents feature names from the data we want to plot on X and Y axis  
def seg(str_x, str_y, clusters):
    x = []
    y = []
    
    for i in range(clusters):
        x.append(data1[str_x][data1['cluster']==i])
        y.append(data1[str_y][data1['cluster']==i])
    
    return x,y

def plot_cluster(str_x, str_y, clusters):
    plt.figure(figsize = (5,5), dpi = 120)
    
    x,y = seg(str_x, str_y, clusters)
    
    for i in range(clusters):
        plt.scatter(x[i], y[i], label = 'cluster{}'.format(i))
        
    plt.xlabel(str_x)
    plt.ylabel(str_y)
    plt.title(str(str_x+' Vs '+str_y))
    plt.legend()
```

```python
plot_cluster('Indians', 'Foreigners', 3)
```

```python
plot_cluster('Indian_Male', 'Foreigners_Male',3)
```

```python
plot_cluster('Indian_Female', 'Foreigners_Female',3)
```

```python
plot_cluster('other', 'Indians',3)
```
